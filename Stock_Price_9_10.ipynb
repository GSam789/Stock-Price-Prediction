```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
adobe_stock_price = pd.read_csv('ADBE.csv')
adobe_stock_price
```

```python
costco_stock_price = pd.read_csv('COST.csv')
costco_stock_price
```

```python
### Check missing values for two stock price datasets
print('Adobe Stock Price:')
print(adobe_stock_price.isnull().sum())
print('\n')

print('Costco Stock Price:')
print(costco_stock_price.isnull().sum())
```

```python
### Create line plot of stock price for two stocks

plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(adobe_stock_price['Date']), adobe_stock_price['Adj Close'], label='Adj Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Adj Close Price($)')
plt.title('Adjsuted Close Price Of Adobe Inc(ADBE) In Recent Five Years')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(pd.to_datetime(costco_stock_price['Date']), costco_stock_price['Adj Close'], label='Adj Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Adj Close Price($)')
plt.title('Adjusted Close Price Of Costco Wholesale Corp(COST) In Recent Five Years')
plt.legend()
plt.show()
```

```python
### Create box-plot of two stocks' close price
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (12, 6))

axes[0].boxplot(adobe_stock_price['Adj Close'], vert = True)
axes[0].set_xlabel('Adobe Inc(ADBE)')
axes[0].set_ylabel('Adj Close Price')
axes[0].set_title('Box Plot of Adobe Adjusted Close Price')

axes[1].boxplot(costco_stock_price['Adj Close'], vert = True)
axes[1].set_xlabel('Costco Wholesale Corp(COST)')
axes[1].set_ylabel('Adj Close Price')
axes[1].set_title('Box Plot of Costco Adjsuted Close Price')

plt.show()
```
